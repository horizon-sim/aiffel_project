```python
# (1) 데이터 가져오기
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("../train.csv")
df.head()
```

```python
# 필요없는 부분
df.drop(["casual", "registered"], axis=1, inplace=True)
df
```

```python
# (2) datetime 컬럼을 datetime 자료형으로 변환하고 연, 월, 일, 시, 분, 초까지 6가지 컬럼 생성
df["datetime"] = pd.to_datetime(df["datetime"])
df["year"] = df["datetime"].dt.year
df["month"] = df["datetime"].dt.month
df["day"] = df["datetime"].dt.day
df["hour"] = df["datetime"].dt.hour
df["minute"] = df["datetime"].dt.minute
df["second"] = df["datetime"].dt.second
df
```

```python
# (3) year, month, day, hour, minute, second 데이터 개수 시각화하기
plt.subplot(3, 2, 1)
sns.countplot(x=df["year"])
plt.title("Year")
plt.ylim([5300, 5500])

plt.subplot(3, 2, 2)
sns.countplot(x=df["month"])
plt.title("Month")

plt.subplot(3, 2, 3)
sns.countplot(x=df["day"])
plt.title("Day")

plt.subplot(3, 2, 4)
sns.countplot(x=df["hour"])
plt.title("Hour")
plt.ylim([300, 600])

plt.subplot(3, 2, 5)
sns.countplot(x=df["minute"])
plt.title("Minute")

plt.subplot(3, 2, 6)
sns.countplot(x=df["second"])
plt.title("Second")

# 그래프 간 간격 조정
plt.tight_layout()

# 그래프 보여주기
plt.show()
```

```python
# (4) X, y train/test
# casual, registered 컬럼은 이미 제거했고 분 초도 계속 0으로 나오니 그냥 제거하도록 하겠습니다.
df.drop(["datetime", "minute", "second"], axis=1, inplace=True)
df
```

```python
X = df.drop(columns=["count"], axis=1)
y = df["count"]
print(X.shape, y.shape)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
# (5) LinearRegression 모델 학습
from sklearn.linear_model import LinearRegression

model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
# (6) 학습된 모델로 X_test에 대한 예측값 출력 및 손실함수값 계산
from sklearn.metrics import mean_squared_error

y_pred = model.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
mse
rmse = np.sqrt(mse)

print(f"MSE : {mse}")
print(f"RMSE : {rmse}")
```

```python
# (7) x축은 temp 또는 humidity로, y축은 count로 예측 결과 시각화하기
plt.scatter(X_test["temp"], y_test, label="true")
plt.scatter(X_test["temp"], y_pred, label="pred")
plt.legend()
plt.show()
```

```python
plt.scatter(X_test["humidity"], y_test, label="true")
plt.scatter(X_test["humidity"], y_pred, label="pred")
plt.legend()
plt.show()
```
